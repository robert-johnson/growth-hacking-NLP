```python
import pandas as pd

df = pd.read_csv(r"C:\Users\rajjr\PycharmProjects\growth_hacking\chapter_3\dictionary_based_sentiment.tsv", sep="\t")

ratings = list(df["rating"])
reviews = list(df["review"])
scores = list(df["sentiment"])

print(ratings)
```

```python
def categorize_sentiment(score):
    category = -1
    if (score >= 0.2):
        category = 1
    elif (-0.2 < score < 0.2):
        category = 0
        
    return category

def categorize_rating(score):
    category = -1
#     print(score)
    if (1.0 < score < 5.0):
        category = 0
    elif (score > 4.0):
        category = 1
        
    return category
```

```python
sentiment_categories = [categorize_sentiment(e) for e in scores]
print(sentiment_categories)

rating_categories = [categorize_rating(e) for e in ratings]
print(rating_categories)
```

```python
from collections import Counter
sentiment_counts = Counter(sentiment_categories)
print(sentiment_counts)

rating_counts = Counter(rating_categories)
print(rating_counts)
```

```python
df['sentiment_category'] = sentiment_categories
df['rating_category'] = rating_categories

print(df.head())
```

```python
with open(r"C:\Users\rajjr\PycharmProjects\growth_hacking\chapter_3\categories.tsv", "w") as outfile:
    outfile.write(df.to_csv(index=False, sep="\t"))
```

```python
from sklearn.metrics import accuracy_score, average_precision_score, f1_score
# calculate the accuracy
accuracy_score(rating_categories, sentiment_categories)
```
